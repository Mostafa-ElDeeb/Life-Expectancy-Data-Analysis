```python
# Creating user-defined function to calculate outlier percentage
def outlier_percent(s):
  s=s.dropna()
  q1 = s.quantile(0.25)
  q3= s.quantile(0.75)
  iqr = q3-q1
  lower = q1-(1.5*iqr)
  upper = q3+(1.5*iqr)
  outlier = s[(s<lower)|(s>upper)]
  return round((len(outlier)/len(s))*100,2)
```

```python
# Using the outlier_percent function to detect outliers percentage in each column
# Note all columns in the provided list are numeric
outlier_report = {col:outlier_percent(life[col])
                    for col in null_col_list}
outlier_report
```

```python
# Creating a function to clean year column based on what we observed
def y_fill(s):
  for i in range(len(s)):
    if i==0:
      if pd.isna(s.iloc[i]):
        s.iloc[i]=s.iloc[i+1]+1
    elif i==max(range(len(s))):
      if pd.isna(s.iloc[i]):
        s.iloc[i]=s.iloc[i-1]-1
    else:
      if pd.isna(s.iloc[i]) and pd.isna(s.iloc[i+1]):
        s.iloc[i]=s.iloc[i-1]-1
      elif pd.isna(s.iloc[i]):
        s.iloc[i]=np.mean([s.iloc[i-1],s.iloc[i+1]])
  return s
```

```python
# Clean year column using the function we have created
life['year']=life.groupby('country')['year'].transform(y_fill)
```

```python
# Studying correlation between columns that represent socioeconomic status
num = life.select_dtypes(include='number')
corr_col = ['population','alcohol','gdp','schooling','income_composition_of_resources','total_expenditure']
num.corr().loc[corr_col, corr_col]
```

```python
# Studying correlation between columns that describe immunizations
corr_imm = ['polio','diphtheria']
num.corr().loc[corr_imm,'hepatitis_b']
```

```python
# Dropping columns that have weak correlation and will not be efficiently imputed using KNN imputation and will not affect our analysis
life.drop(columns=['population','total_expenditure'],inplace=True)
```

```python
# Impute using KNN

# Import necessary tools
from sklearn.impute import KNNImputer
from sklearn.preprocessing import StandardScaler

# Prepare list of columns to be imputed in 2 lists as we discussed before after studying correlation
socio_econ = ['alcohol','gdp','schooling','income_composition_of_resources']
immunz = ['polio','diphtheria','hepatitis_b']

# Standardizing numbers in columns to help the model impute correctly
scaling_se = StandardScaler()
socio_econ_scaled = scaling_se.fit_transform(life[socio_econ])
scaling_im = StandardScaler()
immunz_scaled = scaling_im.fit_transform(life[immunz])

# Impute the columns using KNN
imput = KNNImputer(n_neighbors=5)
socio_econ_imputed = imput.fit_transform(socio_econ_scaled)
immunz_imputed = imput.fit_transform(immunz_scaled)

# Inverse standardized columns to get back to their orginal values
life[socio_econ]=scaling_se.inverse_transform(socio_econ_imputed)
life[immunz]= scaling_im.inverse_transform(immunz_imputed)
```
